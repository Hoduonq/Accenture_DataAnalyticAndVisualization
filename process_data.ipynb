```python
import pandas as pd
```

```python
df1 = pd.read_csv("C:/university/personal project/The Forage/data/Content.csv")
df2 = pd.read_csv("C:/university/personal project/The Forage/data/Reactions.csv")
df3 = pd.read_csv("C:/university/personal project/The Forage/data/ReactionTypes.csv")
```

```python
df1 = df1.dropna()
df1
```

```python
df2 = df2.dropna()
df2
```

```python
df3 = df3.dropna()
df3
```

```python
df1 = df1.drop(["Unnamed: 0", "User ID", "URL"], axis=1)
```

```python
df1
```

```python
df2 = df2.drop(["Unnamed: 0", "User ID"], axis=1)
```

```python
df2
```

```python
df3 = df3.drop("Unnamed: 0", axis=1)
```

```python
df3
```

```python
df1.to_csv('Content_cleaned.csv', index=False)
df2.to_csv('Reactions_cleaned.csv', index=False)
df3.to_csv('ReactionTypes_cleaned.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = df1.merge(df1, how='left', on='Content ID')
data = data.merge(df2, how='left', on='Content ID')
data = data.merge(df3, how='left', on='Type')
data = data.rename(columns={'Type':'React_Type'})
```

```python
data
```

```python
data = data.drop(['Type_x', 'Category_x'], axis=1)
```

```python
data = data.rename(columns={'Type_y':'Type', 'Category_y':'Category'})
```

```python
data
```

```python
data = data.dropna()
```

```python
data = data.drop_duplicates()
```

```python
data
```

```python
data.groupby('Category').agg({"Score" : "sum"}).reset_index().sort_values(by='Score', ascending=False).head()
```

```python
data.to_csv('data.csv', index=False)
```
